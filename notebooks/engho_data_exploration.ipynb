# Exploración de las bases de la ENGHo

En este notebook voy a explorar la ENGHo de manera descriptiva. Son descargables desde (la página del INDEC)[https://www.indec.gob.ar/indec/web/Institucional-Indec-BasesDeDatos-4].

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Carga de datos
### Carga de la base hogares `engho2018_hogares.txt`

```python
hogares_df = pd.read_csv(
    "../data/engho2018_hogares.txt",
    sep="|",
    decimal=".",
    quotechar="\""
)
hogares_df.info()
hogares_df.tail()
```

```python
hogares_cols = [
    "id",           # Clave de identificación del hogar
    "pondera",      # Factor de expansión
    "jniveled",     # Nivel educativo del jefe o la jefa del hogar
    "jocupengh",    # Categoría ocupacional del jefe o la jefa del hogar
    "cantmiem",     # Tamaño del hogar
    "cv1c04",       # Tipo de vivienda
    "cantping",     # Cantidad de perceptores de ingreso
    "reldep",       # Cantidad de miembros por cada miembro perceptor de ingresos
    "regten",       # Régimen de tenencia de la vivienda
    "gastot",       # Gasto de consumo total del hogar
    "gc_01",        # Alimentos y bebidas no alcohólicas
    "gc_02",        # Bebidas alcohólicas y tabaco
    "gc_03",        # Prendas de vestir y calzado
    "gc_04",        # Vivienda, agua, electricidad, gas y otros combustibles
    "gc_05",        # Equipamiento y mantenimiento del hogar
    "gc_06",        # Salud
    "gc_07",        # Transporte
    "gc_08",        # Comunicaciones
    "gc_09",        # Recreación y cultura
    "gc_10",        # Educación
    "gc_11",        # Restaurantes y hoteles
    "gc_12",        # Bienes y servicios varios
    "ingtoth",      # Ingreso total del hogar
    "decgaphr",     # Decil de gasto de consumo per cápita del hogar. Región
    "decgapht",     # Decil de gasto de consumo per cápita del hogar. Total del país
    "hacina",       # Cantidad de miembros por habitación de uso exclusivo
    "ch07",         # Servicios Públicos: Electricidad
    "ch09",         # Servicios Públicos: Agua
    "ch14",         # Servicios Públicos: Cloaca
    "ch15",         # Servicios Públicos: Gas
    "ch21",         # Hogar con aire acondicionado
    "regten",       # Régimen de tenencia de la vivienda
    "propauto"      # Hogar propietario de automóvil
]
```

```python
hog = hogares_df[hogares_cols].copy()
hog.info()
```

### Carga de la base hogares `engho2018_personas.txt`

```python
personas_df = pd.read_csv(
    "../data/engho2018_personas.txt",
    sep="|",
    decimal=".",
    quotechar="\"",
    low_memory=False
)
personas_df.info()
personas_df.tail()
```

```python
personas_cols = [
    "id",           # Clave de identificación del hogar
    "miembro",      # N° de miembro en el hogar
    "pondera",      # Factor de expansión
    "cp03",         # Edad
    "cp04",         # Relación de parentesco con el jefe o la jefa del hogar
    "cp13",         # Sexo
    "region",       # Código de región
    "subregion",    # Código de subregión
    "provincia",    # Código de jurisdicción
    "niveled",      # Nivel educativo
    "cp23_08",      # ¿No cuenta con cobertura médica? 1: No, 2: Sí
    "cat_ocup",     # Categoría ocupacional (EPH)
    "rama_op",      # Rama de actividad de la ocupación principal (CAES)
    "rama_op_2",    # Rama de actividad de la ocupación principal (CAES - letra)
    "cp49",         # ¿Cuántas personas trabajan en ese establecimiento?
    "cp53",         # ¿Cuántas personas trabajan en ese establecimiento?
    "np50"          # En esta ocupación, ¿cuántas horas trabaja habitualmente por semana?
]
```

```python
pers = personas_df[personas_cols].copy()
pers.info()
```
